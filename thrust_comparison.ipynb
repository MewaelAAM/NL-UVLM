```python
import numpy as np
import matplotlib.pyplot as plt

def plot_thrust_comparison(rpms=[4800, 5400, 6000], seed=42):
    np.random.seed(seed)
    
    # Generate data
    base_thrusts = np.array(rpms) / 2000
    experimental = [1.76, 2.14, 2.82]
    rvpm = [1.64, 2.08 , 2.57]
    uvlm = [1.82, 2.43, 3.01]
    
    # Set width of bars and positions of the bars
    barWidth = 0.25
    r1 = np.arange(len(rpms))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    
    # Create the bars
    plt.figure(figsize=(10, 6))
    plt.bar(r1, experimental, width=barWidth, label='Experimental', color='#f65058FF')
    plt.bar(r2, rvpm, width=barWidth, label='rVPM', color='#FBDE44FF')
    plt.bar(r3, uvlm, width=barWidth, label='UVLM', color='#28334AFF')
    
    # Add labels and title with larger font sizes
    plt.xlabel('Motor speed (RPM)', fontsize=16)
    plt.ylabel('Thrust (N)', fontsize=16)
    
    # Increase tick size
    plt.xticks([r + barWidth for r in range(len(rpms))], rpms, fontsize=14)
    plt.yticks(fontsize=14)
    
    # Create legend with larger font
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=16)
    
    # Adjust layout to prevent legend cutoff
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Run the function
plot_thrust_comparison()
```
